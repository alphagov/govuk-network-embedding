```python
urls_to_find = []
cid_name = dict(zip(edges.src_cid, edges.src_node))
for cid,name in zip(edges.dest_cid, edges.dest_node):
    if cid not in cid_name.keys():
        cid_name[cid] = name
urls_to_find =[cid_name[m] for m in still_missing] 
len(urls_to_find)
import gzip
with gzip.open(os.path.join(content_api, "missing_run_3.csv.gz"), "wb") as writer:
    writer.write("Node\n".encode())
    for url in urls_to_find:
        writer.write("{}\n".format(url).encode())
```

```python
path_id = {}
id_count = 0
for path in set(list(edges.src_node)+ list(edges.dest_node)):
    if path not in path_id.keys():
        path_id[path] = id_count
        id_count+=1
```

```python
nid_cid = dict(zip(edges['source'],edges['src_cid']))
for nid,cid in zip(edges['target'],edges['dest_cid']):
    if nid not in nid_cid.keys():
        nid_cid[nid] = cid
```

```python
rowlist = []
index = []
for nid in set(list(edges['source']) + list(edges['target'])):
    row_dict = {}
    row_dict["content_id"] = nid_cid[nid]
    for i,emb in enumerate(cid_emb[nid_cid[nid]]):
        row_dict[i] = emb
    rowlist.append(row_dict)  
    index.append(nid)
```
